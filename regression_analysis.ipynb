```python
import pandas as pd
import matplotlib.pyplot as plt
plt.style.use('ggplot')
from sklearn import datasets
from sklearn import linear_model
import numpy as np
```

```python
# Load dataset
bostonData = datasets.load_boston() #% built-in dataset
yb = bostonData.target.reshape(-1, 1) #getting the target and reshaping the array
Xb = bostonData['data'][:,5].reshape(-1, 1) #geting the column 5 and reshaping the array
```

```python
df = pd.DataFrame(bostonData.data)
df.head()
```

```python
# Plot the variables
plt.scatter(Xb,yb)
plt.ylabel('value of house /1000 ($)')
plt.xlabel('number of rooms')
plt.show()
```

```python
# Create the model
regr = linear_model.LinearRegression()
# Train the model
regr.fit( Xb, yb)
```

```python
# Generate plots
plt.scatter(Xb, yb, color='black')
plt.plot(Xb, regr.predict(Xb), color='blue', linewidth=3)
plt.show()
```
